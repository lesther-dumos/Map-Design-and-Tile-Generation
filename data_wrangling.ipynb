```python
import pandas as pd
```

```python
# Inspecting Data
collisions_df = pd.read_csv("assets\SDOT_Collisions_All_Years.csv")
collisions_df.columns
```

```python
collisions_df.head(2)
```

```python
# Filtering Data
collisions_df_filter = collisions_df[["OBJECTID","X", "Y", "ADDDTTM"]].copy()
collisions_df_filter["Date"] = pd.to_datetime(collisions_df_filter['ADDDTTM']).dt.year

collisions_df_filter = collisions_df_filter[collisions_df_filter['Date'] == 2024]
collisions_df_filter = collisions_df_filter[["OBJECTID","X", "Y"]]
collisions_df_filter.head(100)
```

```python
# Download Data
collisions_df_filter.to_csv('filtered_data_2024_v2.csv', index=False)
```
